```python
from selenium.webdriver.common.by import By

# 페이지 열기
driver.get('http://kartprice.net/')

# CSS 선택자로 요소 찾기
driver.find_element(By.CSS_SELECTOR, 'body > div.pye_wrap > div.pye_header > div.pye_header_m > div > ul > li:nth-child(2) > a').click()
time.sleep(3)

# XPath로 요소 찾기
id_s = driver.find_element(By.XPATH, '//*[@id="id"]')
```

```python
# search_list = [
# '김환기', '이우환', '박서보', '정상화', '김창열', '박수근', '천경자', '이중섭', '윤형근', '이대원', 
# '유영국', '장욱진', '김종학', '백남준', '오치균', '하종현', '도상봉', '이강소', '이왈종', '정선', 
# '김홍도', '김태호', '이상범', '남관', '이응노', '전광영', '김정희', '서세옥', '이성자', '김기창']
search_list = [
'이상범', '남관', '이응노', '전광영', '김정희', '서세옥', '이성자', '김기창']
```

```python
def search(search_word):
    #검색할 칸을 찾아
    search = driver.find_element(By.XPATH,'//*[@id="search_val"]')
    search.clear() # 같은 사이트 여러번 크롤링했을 때의 문제 해결위함
    search.send_keys(search_word)
    time.sleep(1)
    search.send_keys(Keys.ENTER) # 엔터키를 눌러 검색 진행
```

```python
total_title = []
total_author = []
total_year = []
total_standard = []
total_genre = []
total_technique = []
total_source = []
total_date = []
total_price = []
```

## 크롤링 진행

```python
driver.get('http://kartprice.net/')

# 아티스트 목록에 대해 검색 및 크롤링
for artist in search_list:
    search(artist)
    successful_bids = int(driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div/div/div/div[3]/dl/dd/p[2]').text[5:])
    artist_count = 0
    if successful_bids <= 10:
        total_length = 2
    else:
        total_length = successful_bids // 10 + 2

    for j in range(1, total_length):
        page_length = len(driver.find_elements(By.CSS_SELECTOR,
                                               'body > div.pye_wrap > div:nth-child(9) > div > div > table > thead > tr'))
        page_pictures = len(driver.find_elements(By.CSS_SELECTOR, '.money'))
        print('page_pictures:', page_pictures)
        
        for picture in range(2, page_pictures + 1):  ## 게시물 크롤링
            if j == 1:
                child_param = 9
                div_param = 4
            else:
                child_param = 7
                div_param = 3

            title = driver.find_element(By.XPATH,
                                        f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[3]/p")
            author = driver.find_element(By.XPATH,
                                         f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[4]/p")
            year = driver.find_element(By.XPATH,
                                       f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[5]/p[1]")
            standard = driver.find_element(By.XPATH,
                                           f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[5]/p[2]")
            genre = driver.find_element(By.XPATH,
                                        f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[6]/p[1]")
            technique = driver.find_element(By.XPATH,
                                            f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[6]/p[2]")
            source = driver.find_element(By.XPATH,
                                         f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[7]/p[1]")
            date = driver.find_element(By.XPATH,
                                       f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[7]/p[2]")
            price = driver.find_element(By.XPATH,
                                        f"/html/body/div[1]/div[{div_param}]/div/div/table/thead/tr[{picture}]/td[8]")
            
            time.sleep(1)
            total_title.append(title.text)
            total_author.append(author.text)
            total_year.append(year.text)
            total_standard.append(standard.text)
            total_genre.append(genre.text)
            total_technique.append(technique.text)
            total_source.append(source.text)
            total_date.append(date.text)
            total_price.append(price.text)

            # 마지막 페이지에서 크롤링 관련 처리
            if picture == page_pictures:
                try:
                    driver.find_element(By.CSS_SELECTOR, "#paging_t1 > button.next").click()
                    picture = 2
                    time.sleep(2)
                except:
                    driver.get('http://kartprice.net/')
                    time.sleep(2)
                    break

    print(f'{artist} 크롤링 완료, 총 {len(total_title)}건의 크롤링 완료')
```
